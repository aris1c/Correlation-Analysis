#Library

```python
import numpy as np #Library untuk matematika
import pandas as pd #Library untuk dataframe
import seaborn as sns #Library untuk visualisasi
import matplotlib.pyplot as plt
```

Load Data

```python
dataset = sns.load_dataset("mpg")
dataset.head()
```

#Membuat Scatterplot


```python
sns.scatterplot(data = dataset,
                x = "weight",
                y = "mpg")
plt.show() #-0,83 kalo di liat dari heatmap maka ini termasuk kedalam high correlation negative karena mendekati -1
```

```python
sns.scatterplot(data = dataset,
                x = "acceleration",
                y = "mpg")
plt.show()#termasuk ke low correlation
```

```python
correlation_matrix = dataset.corr()  # Matriks korelasi dari dataset

plt.figure(figsize=(10, 8))  # Ukuran gambar

# Buat heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)

plt.title("Heatmap Korelasi")  # Judul gambar
plt.show()
```

```python
sns.scatterplot (data = dataset,
                 x = 'displacement',
                 y = 'mpg')
plt.show() #secara heatmap ini ada di -0,80 dimana ini mendekati -1 maka termasuk high correlation negative
```

```python
sns.scatterplot (data = dataset,
                 x = 'mpg',
                 y = 'mpg')
plt.show() #menunjukkan very high correlation positive
```

```python
sns.scatterplot (data = dataset,
                 x = "mpg",
                 y = "horsepower")
plt.show() #heatmaps masuk di -0,78 atau moderate correlation negative
```

```python
sns.scatterplot(data = dataset,
                x = "displacement",
                y = "horsepower")
plt.show() #0,89 secara correlation heatmaps maka ini termasuk ke high correlation positive
```

#Korelasi Pearson

```python
korelasi = dataset.corr()
korelasi
```

```python
sns.heatmap(korelasi, vmax= 1,vmin = -1,annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5,center = 0)
plt.title("Heatmap korelasi")
plt.show()
```

#Uji Hipotesis

```python
from scipy.stats.stats import pearsonr
```

```python
coef, p_val = pearsonr(dataset["mpg"],dataset["displacement"])
```
